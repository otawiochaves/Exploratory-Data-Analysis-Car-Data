When representing categorical variables, a bar plot is a common approach, but it's important to note that it is not the only option. In this dataset, we encounter variables like "Model" with 915 unique values, which can lead to cluttered visualizations when using a bar graph to display the frequency of each value. To avoid this, filters can be applied to focus only on the values that are of particular interest to the business.

To address this, I will create bar charts specifically for the columns with fewer unique values to ensure a clearer visualization. This will allow us to better understand the distribution and frequency of these categorical variables.

```python
# Select the categorical variables.
cat_vars = ['Transmission Type', 'Driven Wheels','Number of Doors', 'Vehicle Size', 'Vehicle Style']

# Create subplots for each categorical variable
fig, axs = plt.subplots(nrows=len(cat_vars), figsize=(15, len(cat_vars) * 3))
fig.suptitle('Categorical Variables', fontsize=22, y=1.03) # Title.

# Loop through each categorical variable and create a count plot.
for i, var in enumerate(cat_vars):
    ax = axs[i]
    sns.countplot(data=df_categorical, x=var, ax=ax)  # Create count plot.
    ax.set_xlabel(var, fontsize=16)  # Set x-axis label.
    ax.set_ylabel('Count', fontsize=16)  # Set y-axis label.
    sns.despine()

plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees.
plt.tight_layout()
plt.show()
```

<h3 style="text-align:center;font-weight:bold">EDA/ Data Visualizations/ Relationships between variables</h3>

The Pearson correlation heat map is a graphical representation of the correlation between numerical variables in a matrix format. It uses a color scale to show the strength and direction of the relationship between variables. This visualization is particularly useful for identifying linear relationships between variables and selecting relevant variables for further analysis.

*Correlation is typically recommended for numerical variables because it measures the linear relationship between two variables. Numerical variables have a natural ordering and can be represented as continuous values. On the other hand, categorical variables represent distinct categories or groups without any inherent numerical relationship. Calculating correlation between categorical variables may not provide meaningful insights as there is no inherent numerical relationship to measure.* 

```python
# Correlation of numerical variables.
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(df_numeric.corr(method="pearson"), dtype=bool))
heatmap = sns.heatmap(df_numeric.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Numerical variables Pearson Correlation', fontdict={'fontsize': 18}, pad=16);
```

Here are five relevant points from the correlation matrix:

* Engine HP and Engine Cylinders have a strong positive correlation of 0.79, indicating that as the engine horsepower increases, the number of cylinders tends to increase as well.


* City MPG and Highway MPG have a high positive correlation of 0.89, suggesting that vehicles with higher city MPG also tend to have higher highway MPG.


* Engine HP and MSRP (Manufacturer's Suggested Retail Price) show a moderate positive correlation of 0.66, indicating that vehicles with higher horsepower tend to have higher prices.


* Engine Cylinders and Highway MPG exhibit a negative correlation of -0.62, suggesting that vehicles with more cylinders tend to have lower highway MPG.


* Year and Engine HP have a positive correlation of 0.34, implying that newer vehicles generally have higher engine horsepower.
